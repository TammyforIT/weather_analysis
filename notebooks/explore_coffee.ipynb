```python
# TODO: BEGIN YOUR EDA!

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# Print first 5 rows
print(coffee.head())

# Print descriptive statistics
print(coffee.describe())

# Print number of unique values
print(coffee.nunique())

# Print value counts for each column
for col in coffee.columns:
    print(f"Value counts for {col}:")
    print(coffee[col].value_counts())

# Check for null values
null_values = coffee.isnull().sum()
print(null_values)

# Drop null values
coffee = coffee.dropna()

# Convert bearing_trees to numeric
coffee['bearing_trees'] = pd.to_numeric(coffee['bearing_trees'], errors='coerce')

# Filter data for Minas Gerais
minas_gerais = coffee[coffee['subdivision'].str.lower() == 'minas gerais']

# Check if bearing trees are increasing or decreasing
start = minas_gerais['bearing_trees'].values[0]
end = minas_gerais['bearing_trees'].values[-1]
print("Start:", start)
print("End:", end)
if end > start:
    print("Bearing trees are increasing.")
else:
    print("Bearing trees are decreasing or stable.")

# Plot histogram
plt.figure(figsize=(10,6))
sns.histplot(minas_gerais['bearing_trees'], kde=True)
plt.title('Distribution of Bearing Trees in Minas Gerais')
plt.show()

# Create sample dataframe
data = {
    'year': [2002, 2003, 2004, 2005, 2006],
    'bags': [16.2, 17.5, 18.1, 19.2, 20.5]
}
df = pd.DataFrame(data)

# Calculate total bags
total_bags = df['bags'].sum()
print("Total bags:", total_bags)

# Check if production is increasing or decreasing
if df['bags'].iloc[-1] > df['bags'].iloc[0]:
    print("Coffee production is increasing.")
else:
    print("Coffee production is decreasing or stable.")

# Print descriptive statistics
print(df['bags'].describe())

# Calculate correlation coefficients
print("Correlation between bearing trees and 60kgs bag:", coffee['bearing_trees'].corr(coffee['60kgs_bag']))
print("Correlation between bearing hectares and 60kgs bag:", coffee['bearing_hectares'].corr(coffee['60kgs_bag']))
```
